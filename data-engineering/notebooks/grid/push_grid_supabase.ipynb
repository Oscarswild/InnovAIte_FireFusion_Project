# Location Registry Upload

Uploads the **Victoria 1km Grid** coordinate data into the `location_registry` table on Supabase. The CSV is pre-processed to align with the table schema before insertion.

## Imports

```python
import os
from dotenv import load_dotenv
from supabase import create_client
import pandas as pd
import numpy as np
import time
```

## Supabase Client

```python
load_dotenv()

# Get data from .env (ensure this is not pushed to git)
supabase = create_client(
    os.getenv("SUPABASE_URL"),
    os.getenv("SUPABASE_KEY")
)
```

## Load Data

Reads the pre-built 1km grid CSV and renames columns to match the `location_registry` schema. `NaN` values are replaced with `None` for Supabase JSON compatibility.

```python
df = pd.read_csv("../../data/grid/victoria_grid_1km.csv")
df = df.rename(columns={
    "lat": "grid_latitude",
    "lon": "grid_longitude",
    "postcode": "region_name"
})
df = df.replace({np.nan: None})
df
```
